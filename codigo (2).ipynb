# 2.Concatenar  
el siguiente paso sera concatenar para tener todos los datos en un solo archivo  

para concatenar todos los datos primero debemos unir T1 con T2 y luego T3 con T4, así obtenemos dos tablas que despues podremos unir y tener todos los datos en una sola. Esto lo haremos usando la herramienta **.merge()**

```python
T12= pd.merge(T1,T2)
T34= pd.merge(T3,T4)
```

```python
#unimos las dos tablas
T=pd.merge(T12,T34)
```

# Resamplear  
para poder usar la herramienta **.resample()** es necesario transformar los datos en funcion datatime

```python
T['Datetime']= pd.to_datetime(T['Datetime'])
T= T.set_index(T['Datetime'])
```

## Datos en escala de días

```python
Dias= T.resample('D').mean()
Dias
```

## Escala de meses

```python
meses= T.resample('M').mean()
meses
```
